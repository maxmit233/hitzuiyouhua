```python
import skvideo.io
import numpy as np
```

```python
orign = skvideo.io.vread("./data/Video_240p.mp4",as_grey=True)
back = skvideo.io.vread("./data/back.mp4",as_grey=True)
front = skvideo.io.vread("./data/front.mp4",as_grey=True)
front = np.abs(front)
```

```python
import matplotlib.pyplot as plt
def showPic(title,ax,img):
    ax.imshow(img)
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
idxs = sorted([5,6,15,68,34,75])
for idx in idxs:
    f,axs = plt.subplots(1,3)
    showPic("orign",axs[0],orign[idx])
    showPic("back",axs[1],back[idx])
    showPic("front",axs[2],front[idx])
```
